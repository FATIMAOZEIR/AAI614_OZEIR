```python
highest_beer_consumption_state = alcohool['Beer'].idxmax()
print(f"The state with the highest beer consumption is: {highest_beer_consumption_state}")
```

```python
lowest_beer_consumption_state = alcohool['Beer'].idxmin()
print(f"The state with the lowest beer consumption is: {lowest_beer_consumption_state}")
```
