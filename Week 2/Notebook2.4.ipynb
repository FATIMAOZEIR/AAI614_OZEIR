```python
import sys
import subprocess
import pkg_resources

required = {'autoviz'}
installed = {pkg.key for pkg in pkg_resources.working_set}
missing = required - installed

if missing:
    python = sys.executable
    subprocess.check_call([python, '-m', 'pip', 'install', *missing], stdout=subprocess.DEVNULL)

import pandas as pd
import numpy as np
from autoviz import AutoViz_Class
import ssl

ssl._create_default_https_context = ssl._create_unverified_context
```

```python
students = pd.read_csv("https://raw.githubusercontent.com/harmanani/AAI614/main/Week%202/students.csv")
```

##### If you don’t have a target variable you can leave it blank:

```python
target_variable = 'GPA'
```

```python
AV = AutoViz_Class()
sep = ','


dft = AV.AutoViz(students, sep, depVar=target_variable, dfte="",
           header=0, verbose=1,
            lowess=False,chart_format='bokeh',max_rows_analyzed=500,max_cols_analyzed=30)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix
correlation_matrix = dft.select_dtypes(include=[np.number]).corr()

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Draw the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of Numerical Variables in dft')
plt.show()
```
