Question 1: Identify all the outliers in the above data.  Justify your answers using objective measures.

Outliers were detected using the IQR method. Specifically, any value less than Q1 − 1.5 × IQR or greater than Q3 + 1.5 × IQR was treated as an outlier. Using this rule, any data points outside the lower and upper limits in each column were classified as outliers.

```python
import seaborn as sns
import matplotlib.pyplot as plt
Q1 = grads["Avg Hours Studies per Week"].quantile(0.25)
Q3 = grads["Avg Hours Studies per Week"].quantile(0.75)
IQR = Q3 - Q1
lower= Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
outliers = grads[
    (grads["Avg Hours Studies per Week"] < lower) |
    (grads["Avg Hours Studies per Week"] > upper)
]
print(outliers)
```

Question 2: There are various data that are missing.  Fill-in the missing data or delete the rows/columns that you think you should delete.  Justify your answer

The dataset has a small number of missing values in the GPA and Sense of Humour (0–5) columns. Because GPA is numerical, I filled its missing values with the mean to keep all records and maintain the overall distribution. For Sense of Humour, which is an ordinal scale, I used the median since it is less affected by extreme values and better reflects a typical score. I did not delete any rows or columns because the amount of missing data is limited, so removing records would unnecessarily reduce the dataset size.

```python
grads.isnull().sum()
```

```python
grads["GPA"].fillna(grads["GPA"].mean(), inplace=True)
grads["Sense of Humour (0-5)"].fillna(grads["Sense of Humour (0-5)"].median(), inplace=True)
grads.isnull().sum()
```

Question 3: Reload the data and fill-in the data using mean method as well as the frequent method.

```python
import pandas as pd

# Reload data
grads = pd.read_csv("https://raw.githubusercontent.com/harmanani/AAI614/main/Week%203/grads.csv")

grads_mean = grads.copy()

grads_mean["GPA"] = grads_mean["GPA"].fillna(grads_mean["GPA"].mean())
grads_mean["Sense of Humour (0-5)"] = grads_mean["Sense of Humour (0-5)"].fillna(
    grads_mean["Sense of Humour (0-5)"].mean()
)

grads_mean.isnull().sum()
```

```python
grads_mode = grads.copy()

grads_mode["GPA"] = grads_mode["GPA"].fillna(grads_mode["GPA"].mode()[0])
grads_mode["Sense of Humour (0-5)"] = grads_mode["Sense of Humour (0-5)"].fillna(
    grads_mode["Sense of Humour (0-5)"].mode()[0]
)

grads_mode.isnull().sum()
```

Exercise II. Run the cell below to create a new dataframe called `df_miss`.  Its first column will contain some missing values.

```python
import pandas as pd
import numpy as np
import random

nrows = 10
ncols = 5

# set a seed for random number generation
np.random.seed(314)
# create an array filled with random data
data = np.array(np.random.rand(nrows, ncols))
# put the data to a pandas dataframe
df_miss = pd.DataFrame(data)
# rename the columns
df_miss.columns = ['col_'+str(ii) for ii in range(ncols)]

# randomly set some values to missing
ix0 = np.random.randint(nrows, size=3)
ix1 = np.random.randint(nrows, size=3)

df_miss['col_0'][ix0] = np.nan
df_miss['col_1'][ix1] = np.nan

print(df_miss)
```

Impute the missing values (NaN) in `col_0` (but not `col_1`) with the median.  Store the values in the dataframe by using the parameter `inplace`.  Print the dataframe.

```python
df_miss['col_0'].fillna(df_miss['col_0'].median(), inplace=True)
print(df_miss)
```

Impute the missing values in `col_1` with value 0.  Store the values in the dataframe by using the parameter `inplace`.  Print the dataframe.

```python
df_miss['col_1'].fillna(0, inplace=True)

print(df_miss)
```
