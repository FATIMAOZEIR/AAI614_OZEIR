## *US Census Income Data Analysis (1994 Dataset)*

## 1. Data Loading and Importing Libraries

```python

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats

import statsmodels.api as sm
import statsmodels.formula.api as smf


pd.set_option('display.max_columns', None)

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"

df = pd.read_csv(url, header=None)

df.columns = [
    "age", "workclass", "fnlwgt", "education",
    "education_num", "marital_status", "occupation",
    "relationship", "race", "sex",
    "capital_gain", "capital_loss",
    "hours_per_week", "native_country", "income"
]
```

## 4. Capital Gain and Capital Loss Analysis

```python
import matplotlib.pyplot as plt

# Capital gain distribution
plt.figure()
df["capital_gain"].hist(bins=50)
plt.title("Distribution of Capital Gain")
plt.xlabel("capital_gain")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
df["capital_loss"].hist(bins=50)
plt.title("Distribution of Capital Loss")
plt.xlabel("capital_loss")
plt.ylabel("Frequency")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create categorical variables
df["capital_gain_category"] = np.where(df["capital_gain"] > 0, " Gain", "No Gain")
df["capital_loss_category"] = np.where(df["capital_loss"] > 0, " Loss", "No Loss")

# Step 2: Print distributions
print("capital Gain Distribution:")
print(df["capital_gain_category"].value_counts())

print("capital Loss Distribution:")
print(df["capital_loss_category"].value_counts())

# Step 3: Visualizations

plt.figure()
df["capital_gain_category"].value_counts().plot(kind="bar")
plt.title("Capital Gain Category Distribution")
plt.ylabel("Count")
plt.show()

plt.figure()
df["capital_loss_category"].value_counts().plot(kind="bar")
plt.title("Capital Loss Category Distribution")
plt.ylabel("Count")
plt.show()
```

## 5. Sampling Weights (fnlwgt) Analysis

## 6. Linear Regression: Hours per Week and Gender

```python
import statsmodels.api as sm

# Convert sex to numeric
df["sex_binary"] = df["sex"].map({" Male": 1, " Female": 0})

#  Define variables
X = df[["sex_binary"]]
y = df["hours_per_week"]

#  Add constant (intercept)
X = sm.add_constant(X)

#  Fit regression model
model = sm.OLS(y, X).fit()

#  Show results
print(model.summary())
```

```python
import statsmodels.api as sm

#  Data Preparation
df["sex_binary"] = df["sex"].map({" Male": 1, " Female": 0})
df["income_binary"] = df["income"].map({" <=50K": 0, " >50K": 1})
```

```python
# Model 1: Sex only
model_sex_only = sm.add_constant(df[["sex_binary"]])
model1 = sm.OLS(df["hours_per_week"], model_sex_only).fit()
print(model1.summary())
```

```python
# Model 2: Sex + education
model_sex_education = sm.add_constant(df[["sex_binary", "education_num"]])
model2 = sm.OLS(df["hours_per_week"], model_sex_education).fit()
print(model2.summary())
```

```python
# Model 3: Sex + education + income
model_full = sm.add_constant(df[["sex_binary", "education_num", "income_binary"]])
model3 = sm.OLS(df["hours_per_week"], model_full).fit()
print(model3.summary())
```

```python
#model comparison

print("R-squared comparison:")
print("Model 1:", model1.rsquared)
print("Model 2:", model2.rsquared)
print("Model 3:", model3.rsquared)
```

## 8. Correlation Analysis

```python
# Calculate the correlation matrix
correlation_matrix = df[["age", "education_num", "hours_per_week"]].corr()

# Display the correlation matrix
print(correlation_matrix)
```

```python
from scipy.stats import pearsonr

# Split dataset by sex
male = df[df["sex"] == " Male"]
female = df[df["sex"] == " Female"]

# Male correlation
r_male, p_male = pearsonr(male["education_num"], male["age"])
print("Male correlation (education vs age):", r_male)
print("Male p-value:", p_male)

# Female correlation
r_female, p_female = pearsonr(female["education_num"], female["age"])
print(" Female correlation (education vs age):", r_female)
print("Female p-value:", p_female)
```

```python
# Compute covariance matrix
cov_matrix = df[["education_num", "hours_per_week"]].cov()

print(cov_matrix)
```
